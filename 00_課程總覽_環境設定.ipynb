## 1 · 檢查環境與函式庫

先確認電腦裝了哪些套件。缺的之後對應章節再裝。

```python
# 概念：用 importlib 動態檢查套件是否已安裝（不用真的 import 一堆）
import importlib   # 標準庫，可用字串名稱載入模組
import sys

print('Python 版本：', sys.version.split()[0])   # 確認 Python 版本

# 本課程會用到的套件清單
libs = [
    'numpy', 'pandas', 'matplotlib', 'sklearn', 'torch',
    'shapely', 'geopandas', 'networkx', 'scipy', 'PIL',
]

# 概念：for 迴圈 + try/except 防呆（缺套件不讓整段崩潰）
for name in libs:
    try:
        importlib.import_module(name)   # 嘗試載入
        print(f'  [OK]   {name}')        # f-string 格式化輸出
    except ImportError:
        print(f'  [缺]   {name}  （之後章節需要時再安裝）')
```

## 2 · 設定 matplotlib 中文字型

matplotlib 預設不含中文，標題 / 標籤會變成方框。
每本要畫圖的 notebook 開頭都放這段。

```python
# 中文字型設定
import os
import matplotlib.pyplot as plt          # 慣例別名 plt
from matplotlib import font_manager as fm

# 概念：try/except 載入字型檔，找不到就略過（跨平台防爆）
for fp in ['NotoSansCJK-Regular.otf', 'msjh.ttc', 'mingliu.ttc']:
    try:
        if os.path.exists(fp):
            fm.fontManager.addfont(fp)   # 把字型檔註冊進 matplotlib
    except Exception:
        pass                             # 失敗就跳過，不中斷

# 設定優先字型清單（Windows 用微軟正黑體 Microsoft JhengHei）
plt.rcParams['font.family'] = ['Microsoft JhengHei', 'Noto Sans CJK TC', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False   # 注意：不設這行，負號會顯示成方框
print('字型設定完成')
```

## 3 · 畫第一張圖（驗證字型）

畫一張有中文的簡單圖，確認環境正常。

```python
import numpy as np   # 數值運算核心套件

# 概念：最小製圖流程 — 建資料 → 建圖 → 畫 → 標註
x = np.linspace(0, 10, 100)   # 0 到 10 之間 100 個等距點
y = np.sin(x)                 # 對每個點取 sin（numpy 向量化）

fig, ax = plt.subplots(figsize=(6, 3))   # 建立畫布與一個子圖（物件式寫法，推薦）
ax.plot(x, y, label='sin(x)')            # 折線圖
ax.set_title('環境測試：正弦曲線')        # 中文標題（驗證字型）
ax.set_xlabel('x 軸')
ax.set_ylabel('y 軸')
ax.legend()                              # 顯示圖例
ax.grid(True)                            # 格線
plt.tight_layout()                       # 自動排版防重疊
plt.show()
# 技巧：看到中文正常、沒有方框，環境就 OK
```
